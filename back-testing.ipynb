```python
df['Sl'] = 100
df['Tp'] = 200
```

```python
def PL(df):
    if (df['Close']-df['Open'])*df['Dir'] > df['Tp']:
        return df['Tp']
    elif (df['Close']-df['Open'])*df['Dir'] > 0 and (df['Close']-df['Open'])*df['Dir'] < df['Tp']:
        return (df['Close']-df['Open'])*df['Dir']
    elif (df['Close']-df['Open'])*df['Dir'] < -df['Sl']:
        return -df['Sl']
    else:
        return (df['Close']-df['Open'])*df['Dir']
```

```python
PandL = []
for i in range (1000):
    df['Dir'] = np.random.choice([1,-1], df.shape[0])
    df['PL'] = df.apply (lambda df: PL(df), axis = 1)
    PandL.append(df['PL'].sum())
```

```python
def BPL(df):
    if df['Dir'] > 0:
        if df['LH'] > 0:
            if df['High'] - df['Open'] > df['Tp']:
                return df['Tp']
            elif df['Open'] - df['Low'] > df['Sl']:
                return -df['Sl']
            else:
                return df['Close'] - df['Open']
        else:
            if df['Open'] - df['Low'] > df['Sl']:
                return -df['Sl']
            elif df['High'] - df['Open'] > df['Tp']:
                return df['Tp']
            else:
                return df['Close'] - df['Open']
    else:
        if df['LH'] < 0:
            if df['Open'] - df['Low'] > df['Tp']:
                return df['Tp']
            elif df['High'] - df['Open'] > df['Sl']:
                return -df['Sl']
            else:
                return df['Open'] - df['Close']
        else: 
            if df['High'] - df['Open'] > df['Sl']:
                return -df['Sl']
            elif df['Open'] - df['Low'] > df['Tp']:
                return df['Tp']
            else:
                return df['Open'] - df['Close']              
```
